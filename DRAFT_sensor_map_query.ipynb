```python
import pandas as pd
import requests

url = 'https://erddap.sensors.ioos.us/erddap/search/advanced.csv?page=1&itemsPerPage=2000&searchFor=dissolved+oxygen+-usgs&protocol=%28ANY%29&cdm_data_type=%28ANY%29&institution=%28ANY%29&ioos_category=%28ANY%29&keywords=%28ANY%29&long_name=%28ANY%29&standard_name=mass_concentration_of_oxygen_in_sea_water&variableName=%28ANY%29&maxLat=&minLon=&maxLon=&minLat=&minTime=&maxTime='

df = pd.read_csv(url)

print('Found %i datasets.' % len(df))

df_out = pd.DataFrame(columns=['latitude (degrees_north)','longitude (degrees_east)','url'])

skip_data = ['homer-dolphin-surface-water-q',
             'nerrs_hudskwq', 
             'nerrs_lksblwq', 
             'nerrs_lksolwq', 
             'nerrs_lkspowq',
             'nerrs_hudscwq',
             'nerrs_gndpcwq'
            ]

for url in df.tabledap:
    data_url = url+'.csvp?latitude%2Clongitude&distinct()'
    request_response = requests.head(data_url)
    status_code = request_response.status_code
    if status_code != 404:
        print(url)
        df1 = pd.read_csv(data_url)
        df1['url'] = url
        df_out = pd.concat([df_out, df1], ignore_index=True)
    else:
        print('skipping', url)

## Think about using nccsv response to collect metadata too

df_out.head()
```
