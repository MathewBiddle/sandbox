# Query a GitHub repository for issues using the GitHub API.

_Author:_ Mathew Biddle

This notebook exemplifies how to query a GitHub repository's collection of issues to conduct an analysis.

The GitHub repository has an issue template that respondents will fill out and we'd like to extract some pieces of that. The information will be used to curtail the workshop according to the needs of our attendees.

Query the github issues for:
* `open`
* tagged with `data help`

<https://github.com/ioos/bio_data_guide/issues?q=is%3Aissue+is%3Aopen+label%3A%22data+help%22>

```python
import pandas as pd
import urllib.parse

state = 'open'

labels = 'data help'

# percent encode
labels = urllib.parse.quote(labels)

# build api call
url = 'https://api.github.com/repos/ioos/bio_data_guide/issues?per_page=10000&state={}&labels={}'.format(state,labels)

# read json response
df = pd.read_json(url)

df
```

Save date strings as datetime objects to do time parsing

```python
df['created_at']= pd.to_datetime(df['created_at'])

df
```

Find issues that were submitted after `2023-01-09` (when we asked for contributions).

Collect those issue body's and split it into another data frame so we can extract the info we want.

```python
df_body = df.loc[df['created_at']>'2023-01-09','body'].str.split('###',expand=True,)

df_body.rename(columns={1: 'contact',
                        2: 'title',
                        3: 'description',
                        4: 'link_to_data'}, 
               inplace=True)

df_body
```

Clean up the description by removing standard text and newlines.

```python
df_body['description'] = df_body['description'].str.replace("Describe your dataset and any specific challenges or blockers you have or anticipate","").str.replace('\r',' ').str.replace('\n',' ')

df_body
```
