```python
import gspread
import pandas as pd
from google.colab import auth
from oauth2client.client import GoogleCredentials
from google.auth import default

creds, _ = default()
## For google spreadsheet reading you need to authenticate w/ google
auth.authenticate_user()
```

```python
def get_sheet_data(url: str, sheet: str) -> pd.DataFrame:
  '''
  Gets data from specific worksheet in Google Spreadsheet and
  returns as a DataFrame.

  parameters:
  sheet (str): name of worksheet, eg 'Form Responses 1'

  returns:
  DataFrame: worksheet data as a DataFrame

  '''

  gc = gspread.authorize(creds)
  workbook = gc.open_by_url(url)
  worksheet = workbook.worksheet(sheet)

  df = pd.DataFrame(worksheet.get_all_records())

  return df
```

```python
url = 'https://docs.google.com/spreadsheets/d/1jBS8ASS27yV8APZ8Fh-tgX6dHdopwianrUZv0kbKcxw/edit?resourcekey#gid=1284796732'
sheet = 'Form Responses 1'

df = get_sheet_data(url, sheet)

df
```

```python
df_out = df.replace('^(?i)yes$',True, regex=True).replace('^(?i)no$',False,regex=True)
df_out
```
