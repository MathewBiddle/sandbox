# Reading 
Results from https://github.com/evsatt/Final_GOOS_BioEco

Credit: Satterthwaite et al. 2021 "Establishing the foundation for the global observing system for marine life".

Read the data from the [GitHub repository](https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv).

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv'

df = pd.read_csv(url)

print('Initial size:\t{}\n'.format(df.shape))

print('Column names:')
for col in df.columns.tolist():
  print('\t'+col)

df.head()
```

# Subsetting

Define the extra columns to return in the response.

```python
extra_cols = ['prog_name','prog_url','resp_firstname','resp_lastname','resp_email']
```

## Marine Mammals

Find column names that contain the string `Mammal`.

```python
marine_mammals_cols = df.filter(regex='Mammal').columns.tolist() # MM?
marine_mammals_cols
```

### Filter

Using the column names from the list above, look for rows where those columns
have a response (not NA). This returns a filtered data frame which contains only
the rows which have entries for the columns specified.

Then, we remove the rows have the text `This variable is not part of the network/program in any way` in the `_notpart` columns.

As we filter the data frame, we print the size of the data frame as a check.

```python
print('Initial size:\t\t\t\t\t{}'.format(df.shape))

df_mm = df.loc[( df['Marine_Mammals'].notna() | 
                       df['Marine_Mammal_abundance_systematic'].notna() |
                       df['Marine_Mammal_abundance_opportunistic'].notna() |
                       df['Marine_Mammal_abundance_aggregate'].notna() |
                       df['Marine_Mammal_abundance_dataproduct'].notna() |
                       df['Marine_Mammal_abundance_datause'].notna() |
                       df['Marine_Mammal_distribution_systematic'].notna() |
                       df['Marine_Mammal_distribution_opportunistic'].notna() |
                       df['Marine_Mammal_distribution_aggregate'].notna() |
                       df['Marine_Mammal_distribution_dataproduct'].notna() |
                       df['Marine_Mammal_distribution_datause'].notna()
                        )]

print('Size after removing all NA seagrass:\t\t{}'.format(df_mm.shape))

df1 = df_mm.drop(df_mm[df_mm['Marine_Mammal_abundance_notpart'].str.contains('This variable is not part of the network/program in any way',na=False)].index)
df2 = df1.drop(df1[df1['Marine_Mammal_distribution_notpart'].str.contains('This variable is not part of the network/program in any way',na=False)].index)

print('Size after removing "not part of network":\t{}'.format(df2.shape))
```

Now save and print out the data of interest.

```python
pd.set_option('max_rows', 99999)
pd.set_option('max_colwidth', 400)

cols = extra_cols + marine_mammals_cols
df2[cols].to_csv('marine_mammals_response.csv')
df2[cols]
```

## Seagrass
Compile a list of all columns that contain `Seagrass`.

```python
seagrass_cols = df.filter(regex='Seagrass').columns.tolist()
seagrass_cols
```

### Filter

Using the column names from the list above, look for rows where those columns
have a response (not NA). This returns a filtered data frame which contains only
the rows which have entries for the columns specified.

Then, we remove the rows have the text `This variable is not part of the network/program in any way` in the `_notpart` columns.

As we filter the data frame, we print the size of the data frame as a check.

```python
print('Initial size:\t\t\t\t\t{}'.format(df.shape))

df_seagrass = df.loc[( df['Seagrass'].notna() | 
                       df['Seagrass_cover_systematic'].notna() |
                       df['Seagrass_cover_opportunistic'].notna() |
                       df['Seagrass_cover_aggregate'].notna() |
                       df['Seagrass_cover_dataproduct'].notna() |
                       df['Seagrass_cover_datause'].notna() |
                       #df['Seagrass_cover_notpart'].notna() |
                       df['Seagrass_composition_systematic'].notna() |
                       df['Seagrass_composition_opportunistic'].notna() |
                       df['Seagrass_composition_aggregate'].notna() |
                       df['Seagrass_composition_dataproduct'].notna() |
                       df['Seagrass_composition_datause'].notna()
                        )]

print('Size after removing all NA seagrass:\t\t{}'.format(df_seagrass.shape))

df1 = df_seagrass.drop(df_seagrass[df_seagrass['Seagrass_composition_notpart'].str.contains('This variable is not part of the network/program in any way',na=False)].index)
df2 = df1.drop(df1[df1['Seagrass_cover_notpart'].str.contains('This variable is not part of the network/program in any way',na=False)].index)

print('Size after removing "not part of network":\t{}'.format(df2.shape))
```

Now save and print out the data of interest.


```python
pd.set_option('max_rows', 99999)
pd.set_option('max_colwidth', 400)

cols = extra_cols + seagrass_cols
df2[cols].to_csv('seagrass_response.csv')
df2[cols]
```
