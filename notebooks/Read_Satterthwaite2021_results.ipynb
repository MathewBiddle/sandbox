# Reading 
Results from https://github.com/evsatt/Final_GOOS_BioEco

Credit: Satterthwaite et al. 2021 "Establishing the foundation for the global observing system for marine life".

Read the data from the [GitHub repository](https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv).

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv'

df = pd.read_csv(url)

print('Initial size:\t{}\n'.format(df.shape))

print('Column names:')
for col in df.columns.tolist():
  print('\t'+col)

df.head()
```

# Subsetting

The function below returns the row indexes where there are useful contents in the columns related to the search term.

```python
def idx_regexp(df_in: pd.DataFrame, term: str):

  regex = '%s(.*)_notpart' % term

  # filter data frame to columns with term
  df = df_in.filter(regex = term)

  # drop rows where all values are NA
  df_nona = df.dropna(how = 'all')

  # Find columns with term and _notpart. 
  # drop the rows that have NA
  # This finds the row index we need to remove from above
  df_notpart = df_in.filter(regex = regex).dropna(how='all')

  # drop rows where _notpart exists
  df_out = df_nona.drop(index=df_notpart.index, axis=0)

  return df_out.index
```

## Marine Mammals

Subset to the rows which reference `Mammal`.

```python
idx = idx_regexp(df, 'Mammal')

df_mammals = df.iloc[idx]

df_mammals.to_csv('marine_mammals_response.csv')
df_mammals
```

## Seagrass
Compile a list of all columns that contain `Seagrass`.

```python
idx = idx_regexp(df, 'Seagrass')

df_seagrass = df.iloc[idx]

df_seagrass.to_csv('seagrass_response.csv')
df_seagrass
```
