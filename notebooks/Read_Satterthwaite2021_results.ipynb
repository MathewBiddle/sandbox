Reading results from https://github.com/evsatt/Final_GOOS_BioEco

Credit: Satterthwaite et al. 2021 "Establishing the foundation for the global observing system for marine life".

```python
import pandas as pd
```

Read the data from https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv

```python
url = 'https://raw.githubusercontent.com/evsatt/Final_GOOS_BioEco/main/Data_Satterthwaite2021/1Final_BiologicalOceanObservingSurveyData_420_8132020_Satterthwaite2021.csv'
df = pd.read_csv(url)

print(df.shape)
df.head()
```

Compile a list of all columns that contain `Seagrass`.

```python
df.filter(regex='Seagrass').columns.tolist()
```

collect all rows that have an entry in one of the columns containing `Seagrass`.

```python
df_seagrass = df.loc[( df['Seagrass'].notna() | 
                       df['Seagrass_cover_systematic'].notna() |
                       df['Seagrass_cover_opportunistic'].notna() |
                       df['Seagrass_cover_aggregate'].notna() |
                       df['Seagrass_cover_dataproduct'].notna() |
                       df['Seagrass_cover_datause'].notna() |
                       df['Seagrass_cover_notpart'].notna() |
                       df['Seagrass_composition_systematic'].notna() |
                       df['Seagrass_composition_opportunistic'].notna() |
                       df['Seagrass_composition_aggregate'].notna() |
                       df['Seagrass_composition_dataproduct'].notna() |
                       df['Seagrass_composition_datause'].notna() |
                       df['Seagrass_composition_notpart'].notna()
                        )]
```

```python
import json

import pprint
#pprint.pprint(json.loads(df_seagrass['spatial_geojson'][0]))
print(json.dumps(json.loads(df_seagrass['spatial_geojson'][0])['features'],indent=4, sort_keys=False))
```

Attempting to concatenate all the json parts together into one geojson.

```python
result = ["{\"type\": \"FeatureCollection\",\"features\": ["]

for entry in df_seagrass.loc[df_seagrass['spatial_geojson'].notna(), 'spatial_geojson']:
  if 'features' in json.loads(entry):
    result.append(json.loads(entry)['features'][0])

result.append("  }]")
#print(result)
#with open("merged_file.json", "wb") as outfile:
#    json.dump(result, outfile)
json.dumps(result, indent=4, sort_keys=False)
```
