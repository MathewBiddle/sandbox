# Testing WMS services from MBON portal

This notebook tests the WMS service from the MBON portal and plots the NOAA Deep Sea Coral database.

MBON Portal link: https://mbon.ioos.us/#metadata/415c622b-d513-4f7e-bf27-a128f54ee68f/bc9ef62e-dc2d-48d6-ab3e-9f3ddae383cf

```python
from owslib.wms import WebMapService
```

```python
url = 'https://data.axds.co/gs/noaa_deepseacorals/wms?SERVICE=WMS&REQUEST=GetMap&LAYERS=noaa_deepseacorals:dsc_natdb&VERSION=1.1.1&FORMAT=image/png'
```

```python
wms = WebMapService(url)
img = wms.getmap(   layers=['noaa_deepseacorals:dsc_natdb'],
                     #styles=['visual_bright'],
                     srs='EPSG:4326',
                     bbox=(-180, -90, 180, 90),
                     size=(600, 500),
                     format='image/jpeg',
                     transparent=True
                     )

out = open('deep_sea_corals.jpg', 'wb')
out.write(img.read())
out.close()
```
